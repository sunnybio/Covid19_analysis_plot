```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid19_df=pd.read_csv("total-deaths-and-cases-covid-19.csv")
covid19_df_china=covid19_df[covid19_df['Entity']=='China']
covid19_df_china=covid19_df_china.reset_index()
covid19_df_china=covid19_df_china[['Date','Total confirmed deaths (deaths)','Total confirmed cases (cases)']]
covid19_df_china.columns=['Date','deaths','cases']

covid19_df_china.head(50)
```

```python
covid19_df_india=covid19_df[covid19_df['Entity']=='India']
covid19_df_india=covid19_df_india.reset_index()
covid19_df_india=covid19_df_india[['Date','Total confirmed deaths (deaths)','Total confirmed cases (cases)']]
covid19_df_india.columns=['Date','deaths','cases']
covid19_df_india.head()
```

```python
sns.set_style('darkgrid')
plt.figure()
plt.plot(covid19_df_china.index.tolist(),covid19_df_china['cases'],label='China(Cases)')
plt.plot(covid19_df_china.index.tolist(),covid19_df_china['deaths'],label='China(deaths)')

plt.plot(covid19_df_india.index.tolist(),covid19_df_india['cases'],label='India(Cases)')
plt.plot(covid19_df_india.index.tolist(),covid19_df_india['deaths'],label='India(deaths)')
ax = plt.gca()

ax.set_xlabel('No. of days')
ax.set_title('Covid19 Usa and India')
sns.set_style('darkgrid')
plt.legend()
```
